```python
def calc_income(df):
    '''
    INPUT: df - > dataframe
    
    OUTPUT: result -> dataframe
    '''
    
    result = df.groupby('gender')['income'].aggregate(["mean"])
    result.columns = ['income']
    
    return result
    
```

```python
calc_income(df)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['John', 'Jane', 'Bob', 'Mary'],
        'Age': [17, 25, 55, 42]}
df = pd.DataFrame(data)

# Define a function to categorize age
def categorize_age(age):
    if age < 18:
        return 'kid'
    elif age >= 18 and age <= 50:
        return 'adult'
    else:
        return 'senior'

# Apply the function to the Age column and create a new column 'AgeCategory'
df['AgeCategory'] = df['Age'].apply(categorize_age)

# Print the resulting DataFrame
print(df)
```

```python
# Calculate survival rate for men and women
total_men = df[df['sex'] == 'male'].shape[0]
total_women = df[df['sex'] == 'female'].shape[0]

men_survived = df[(df['sex'] == 'male') & (df['survived'] == 1)].shape[0]
women_survived = df[(df['sex'] == 'female') & (df['survived'] == 1)].shape[0]

survival_rate_men = (men_survived / total_men) * 100
survival_rate_women = (women_survived / total_women) * 100

print("Survival rate of men: {:.2f}%".format(survival_rate_men))
print("Survival rate of women: {:.2f}%".format(survival_rate_women))
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
# age_range = [0,5,10,15,20,25,30,35,40,45,50,55,60,65,70]
age_range = [0,19,50,100]
df['Age Range'] = pd.cut(df['age'], age_range)

#plot the seaborn graph
sns.barplot(x="Age Range", y="survived", data=df[df['sex']=='female'])
plt.title("Survival Rate of Women in the Age Range of Titanic Dataset")
plt.show()
```

```python
age_range = [0,19,50,100]
df['Age Range'] = pd.cut(df['age'], age_range)

#plot the seaborn graph
sns.countplot(x="Age Range", hue="survived", data=df[df['sex']=='female'])
plt.title("Survival Rate of Women in the Age Range of Titanic Dataset")
plt.show()
```

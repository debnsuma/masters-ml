```python
protein = eval(input())
data = protein

fig = plt.figure()

# Count the number of cereals for different amounts of protein available
unique_proteins, protein_counts = np.unique(data, return_counts=True)

# Sort the protein amounts in ascending order
sorted_indices = np.argsort(unique_proteins)
sorted_proteins = unique_proteins[sorted_indices]
sorted_counts = protein_counts[sorted_indices]

# Create the bar plot
plt.bar(sorted_proteins, sorted_counts, color='g')

# Set x-axis and y-axis labels and title
plt.xlabel('Protein amount in grams')
plt.ylabel('Number of cereals')
plt.title('Protein comparison')

# Show the plot
plt.show()
```

```python
fig = plt.figure()
plt.hist(protein, bins='auto')
# Set x-axis and y-axis labels and title
plt.xlabel('Protein amount in grams')
plt.ylabel('Number of cereals')
plt.title('Protein comparison')
```

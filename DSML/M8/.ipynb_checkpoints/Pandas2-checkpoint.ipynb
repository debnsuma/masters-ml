```python
def solve(df1,df2):
    # just return the 2d list explained in the description
    # Example of an output [9.125, [8.5, 8.666666666666666]]
    
    chem_score = pd.concat((df1.loc[:, ['chem_score']], 
                        df2.loc[:, ['chem_score']]))

    ans1 = np.average(chem_score.chem_score)

    common_students = df1.merge(df2, on='name')
    
    ans2 = list(np.average(common_students.iloc[:, 1:], axis=1))
    
    ans = [ans1, ans2]
    
    return ans
```

```python
def calc_density(df): 
    '''
    INPUT: df -> dataframe
    
    OUTPUT: result -> A series
    '''
    
    ## STEP 1: calculate population density
    
    pop_den = df.loc[:, 'population'] / df.loc[:, 'area'] 
    
    ## STEP 2: sort the values
    
    result = pop_den.sort_values(ascending=False)

    
    return result
```

```python
calc_density(df)
```

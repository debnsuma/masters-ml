```python
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
y=df[['selling_price']]
X=df.drop('selling_price', axis=1)
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_sm = sm.add_constant(X_train)  # Statmodels default is without intercept, to add intercept we need to add constant.

model = sm.OLS(y_train, X_sm)
results = model.fit()

# Print the summary statistics of the model
print(results.summary())
```

```python
results.predict(X_sm)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X_t = pd.DataFrame(X_train, columns=X_train.columns)
vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_new = X_train.drop(['age', 'Diesel'], axis=1)
```

```python
X_sm = sm.add_constant(X_train_new)
model_new = sm.OLS(y_train, X_sm)
results_new = model_new.fit()

print(results_new.summary())
```

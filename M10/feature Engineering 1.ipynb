# Q3. Get a loan


```python
df_graduate_unmarried = df[(df["Married"] == "No") & (df["Gender"] == "Male") & (df['Education'] == "Graduate")]
df_female_graduate = df[(df["Gender"] == "Female") & (df['Education'] == "Graduate")]
```

```python
df_both = pd.concat([df_graduate_unmarried, df_female_graduate])
```

```python
# Create a contingency table
contingency_table = pd.crosstab(
    df_both['Gender'], df_both['Loan_Status']
)
```

```python
contingency_table
```

```python
chi2, p, dof, expected = scipy.stats.chi2_contingency(contingency_table)
```

# Q4. Create new feature

```python
df = pd.read_csv('loan.csv')
df.dropna(inplace=True)
```

```python
df
```

```python
df['NewFeature'] = (df["ApplicantIncome"]) + (3 * df["LoanAmount"]) + (7000 * df["Credit_History"])
```

```python
df['NewFeature']
```

```python
df.describe()
```

```python
threshold  = df['NewFeature'].mean() * 0.25
```

```python
df['Separator'] = (df['NewFeature'] > threshold).astype('int')
```

```python
df.describe(include=object)
```

```python
# Create a contingency table
contingency_table = pd.crosstab(
    df['Loan_Status'], df['Separator']
)
```

```python
contingency_table
```

```python
scipy.stats.chi2_contingency(contingency_table)
```

```python
round(scipy.stats.chi2_contingency(contingency_table)[1], 4)
```

```python
from scipy.stats import poisson

# Define the average rate of value
lambda_val = 499 * (1/365)  # expected number of people who share your birthday

# Calculate the Poisson distribution for k=1
prob_1_birthday = poisson.pmf(1, lambda_val)

print(f"The probability that exactly one other guest has the same birthday as you is {prob_1_birthday}")
```

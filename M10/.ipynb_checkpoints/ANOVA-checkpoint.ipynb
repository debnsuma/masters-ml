# ANOVA

- It says Ho : `All means are EQUAL`
- It says Ha : `All means are DIFFERENT`


```python
sns.kdeplot(hue='Product', x='Income', data=df);
```

# Q1

```python
a = [13, 8, 11, 12, 11]
b = [15, 10, 16, 11, 13, 10]
c = [5, 11, 9, 5]
d = [8, 10, 6, 5, 7]
```

```python
from scipy.stats import f_oneway
```

```python
alpha = 0.05
test_stat, p_value = f_oneway(a, b, c, d)
print(f" test_static = {test_stat} \n p_value = {p_value}")

if p_value < alpha:
    print("reject Ho")
else:
    print("fail to reject ho")
```

# Q2

```python
a = [382, 391, 335, 368, 400, 372]
b = [560, 343, 512, 329, 391, 367]
c = [384, 458, 409, 309, 374, 459]
d = [325, 390, 304, 240, 306, 169]
e = [360, 298, 272, 368, 320, 326]
```

```python
alpha = 0.05
test_stat, p_value = f_oneway(a, b, c, d, e)
print(f" test_static = {test_stat} \n p_value = {p_value}")

if p_value < alpha:
    print("reject Ho")
else:
    print("fail to reject ho")
```

# Q3

# 5. How do I rename columns in a pandas DataFrame? (9:36)

```python
url = 'http://bit.ly/'
ufo = pd.read_csv(url + 'uforeports')
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
ufo.rename(columns={'Colors Reported' : 'Colors_Reported',
                    'Shape Reported' : 'Shape_Reported'},
           inplace=True)
```

```python
ufo.columns
```

```python
new_cols = [i.lower() for i in ufo.columns]
new_cols
```

```python
ufo.columns = new_cols
```

```python
ufo.head()
```

```python
ufo.columns
```

```python
url = 'http://bit.ly/'
ufo = pd.read_csv(url + 'uforeports', names=new_cols, header=None)
```

```python
ufo.head()
```

```python
ufo = pd.read_csv(url + 'uforeports', names=new_cols, header=0)
```

```python
ufo.head()
```

```python
ufo = pd.read_csv(url + 'uforeports')
```

```python
ufo.columns
```

```python
ufo.columns = ufo.columns.str.replace(' ', '_')
```

```python
ufo.head()
```

# 9. How do I apply multiple filter criteria to a pandas DataFrame? (9:51)

```python
movies[(movies['duration'] >= 200) & (movies['star_rating'] > 7)] 
```

```python
movies.loc[ (movies.genre == 'Crime') | (movies.genre == 'Adventure') | (movies.genre == 'Drama'), : ]
```

```python
movies.loc[ movies.genre.isin(['Crime', 'Adventure', 'Drama']) , :]
```

```python
movies.loc[ movies['genre'].isin(['Crime', 'Adventure', 'Drama']) , :]
```

# 14. When should I use a "groupby" in pandas? (8:24)

```python
drinks
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby(['country','continent']).beer_servings.agg(['sum', 'count', 'mean'])
```

```python
%matplotlib inline

drinks.groupby('continent')['beer_servings', 'wine_servings'].mean().plot(kind='bar')
```

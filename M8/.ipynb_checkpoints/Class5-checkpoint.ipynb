## Merging/Stacking 

```python
x = np.arange(5)
```

```python
x
```

```python
np.vstack([x,x,x,x])
```

```python
b = x.reshape(5,1)
```

```python
b
```

```python
np.hstack([b, b, b])
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])

np.hstack((a, b))
```

```python
z = np.array([[2, 4]])
```

```python
z.ndim
```

```python
z.shape
```

```python
np.concatenate([z,z], axis=0)
```

```python
np.vstack([z,z])
```

```python
np.concatenate([z,z])
```

```python
import matplotlib.pyplot as plt 

a = plt.imread('fruits.png')
b = np.concatenate([a,a], axis=1)
plt.imshow(b)
```

```python
b.shape
```

```python
a = plt.imread('fruits.png')
b = np.concatenate([a,a], axis=0)
plt.imshow(b)
```

```python

def get_amount(customers, orders, name):
    '''
    INPUT: 
        customer, orders -> dataframe
        name -> string
        
    OUTPUT: 
        (merged_df, sum_amount) -> tuple of (dataframe, float)
    
    '''
    
    ## STEP 1: merge customer and orders dataframe
    
    merged_df = customers.merge(orders, how='left', on='cust_id')
    
    ## STEP 2: calculate sum of amount for given name
    
    sum_amount = merged_df[merged_df['name'] == 'morty']['amount'].sum(axis=0)
    
    return merged_df, sum_amount
```

```python
customers = pd.DataFrame({'cust_id': [101, 102, 103, 104], 'name': ['rick', 'morty', 'pickle', 'jerry']})
orders = pd.DataFrame({'order_id':['OR1', 'OR3', 'OR23', 'OR42'], 'cust_id' : [102, 105, 101, 102], 'amount': [1200, 650, 120, 989]})
```

```python
get_amount(customers, orders, name='morty')
```

```python
 merged_df = customers.merge(orders, how='left', on='cust_id')
```
